```python

import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap

```

```python
# --- STEP 1: Generate Location Data ---
# Simulating a city with existing stores (Supply) and customer density (Demand)
np.random.seed(42)
num_locations = 100

data = {
    'Latitude': np.random.uniform(12.9, 13.1, num_locations), # Sample coordinates (e.g. Bangalore area)
    'Longitude': np.random.uniform(77.5, 77.7, num_locations),
    'Customer_Demand': np.random.randint(10, 100, num_locations),
    'Existing_Presence': np.random.choice([0, 1], size=num_locations, p=[0.8, 0.2])
}

df_geo = pd.DataFrame(data)

# Calculate 'Expansion Score' (High Demand + Low Presence = High Opportunity)
df_geo['Expansion_Score'] = df_geo['Customer_Demand'] * (1 - df_geo['Existing_Presence'])

print("### GEOSPATIAL DATA PREVIEW ###")
display(df_geo.head())
```

```python
# --- STEP 2: Create Interactive Map ---
# Centering the map on the average coordinates
map_center = [df_geo['Latitude'].mean(), df_geo['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=12, tiles='CartoDB positron')

# 1. Add existing stores as blue markers
for _, row in df_geo[df_geo['Existing_Presence'] == 1].iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color='blue',
        fill=True,
        popup='Existing Store'
    ).add_to(m)

# 2. Add Heatmap of Expansion Opportunity (Expansion_Score)
heat_data = [[row['Latitude'], row['Longitude'], row['Expansion_Score']] for _, row in df_geo.iterrows()]
HeatMap(heat_data, radius=15, blur=10, min_opacity=0.5).add_to(m)
```

```python
# --- STEP 3: Display Results ---
print("\HEATMAP")
display(m)
```

```python
# --- STEP 4: Strategy Insights ---
def expansion_recommendation(df):
    top_spot = df.sort_values(by='Expansion_Score', ascending=False).iloc[0]
    print("\nBUSINESS RECOMMENDATION")
    print(f"Optimal Expansion Coordinate: {top_spot['Latitude']:.4f}, {top_spot['Longitude']:.4f}")
    print(f"Reason: This region shows a peak demand score of {top_spot['Customer_Demand']} with zero current presence.")

expansion_recommendation(df_geo)
```
